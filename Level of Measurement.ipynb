# Levels of Measurement

## . Nominal data -
***The nominal level represents categories that cannot be put in any oreder.***

***eg.Car brands : Maruti ,Honda,Mahindra,Hyundai subjects:Maths,Humantities,Commerece,Biology***

# .Ordinal data-
***ordinal represents categories that can be ordered .
eg Rating of a food -Neutral ,Tasty
Programming level-Beginner,Intermediate,Advances***

# .Interval Data-

Interval data – This level of measurement has no true zero. This means that a zero does not mean absence of a property.
Eg. Temperature in Celsius or Fahrenheit: 0°C or 0°F does not mean the lowest temperature.

Credit scores range from 200–800. Here 0 does not mean true zero.

# .Ratio data –
This level of measurement has a true zero.
Eg. Temperature in Kelvin: 0 K means the lowest possible temperature.
Height of an object – If the height of an object is 0 cm then it means no height.
